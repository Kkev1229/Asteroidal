Importing asteroid module from asteroidal

```python
import asteroidal.asteroid as ast
```

To get started, create an Asteroid object using either the name of the asteroid, number (given by MPC), or Gaia source_id. Not all asteroids might have been observed by Gaia. Easiest way to get an asteroid is to try different asteroid numbers (Hint: No asteroid exists for number_mp < 8).

Example: Create Asteroid object using name of asteroid (flora) and print table of all observations from Gaia

```python
flora = ast.Asteroid(denomination='flora')
print(flora.observations)
```

Plot all observations (ra, dec) as seen by Gaia

```python
flora.plot_observations()
```

Plot the first transit observation

```python
flora.plot_transit(0)
```

Plot all the transits seen by Gaia

```python
flora.plot_all_transits()
```

Plot the orbit of Flora compared to the orbit of planets

```python
flora.plot_orbits()
```
